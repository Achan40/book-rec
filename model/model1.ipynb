```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/books.csv")
data
```

```python
# Calculating the word count for book description
data['word_count'] = data['description'].apply(lambda x: len(str(x).split()))# Plotting the word count
data['word_count'].plot(
    kind='hist',
    bins = 50,
    figsize = (12,8),title='Word Count Distribution for book descriptions')
```
